```python
import pandas as pd
```

```python
_2021 = pd.read_csv('_2021.csv', encoding='utf-8')
_2021['Injeção na rede (kWh)'].fillna("None", inplace=True)

_2022 = pd.read_csv('_2022.csv', encoding='utf-8')
_2022['Injeção na rede (kWh)'].fillna("None", inplace=True)

all = pd.concat([_2021, _2022], ignore_index=True)
all.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
all.info()
```

```python
all.head()
```

```python
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score

from sklearn.model_selection import cross_val_score

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all.drop('Data & Hora', axis=1, inplace=True)
```

```python
#Hot
weather_description = pd.get_dummies (all['weather_description'], drop_first=True)
all.drop('weather_description', axis=1, inplace=True)
all = pd.concat ([all, weather_description], axis=1)

#label
```

```python
all.info()
```

```python
X = all.drop(['Injeção na rede (kWh)'], axis=1)
y = all['Injeção na rede (kWh)'].to_frame()

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.30, random_state=2021)
```

```python
dt = DecisionTreeClassifier (random_state=2022)
dt.fit (X_train, y_train)
```

```python
dt.get_depth()
```

```python
max_depth = dt.get_depth()

param_grid = { 'criterion': ['gini', 'entropy']
               ,'max_depth': [max_depth for max_depth in range (1, max_depth+1)]
               #,'max_depth': [max_depth for max_depth in range (1, 99+1)]
             }

grid = GridSearchCV(dt, param_grid, refit=True, verbose=2)

grid.fit (X_train, y_train)
```

```python
grid.best_params_
```

```python
Predictions = grid.predict (X_test)

print ( classification_report (y_test, Predictions) )
```

```python
ccp_alphas = dt.cost_complexity_pruning_path(X_train, y_train)['ccp_alphas']
```

```python
param_grid = {'ccp_alpha': [alpha for alpha in ccp_alphas]}

grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid, refit=True, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
Predictions_2 = grid_search.predict (X_test)

print ( classification_report (y_test, Predictions_2) )
```

```python
clf_2 = DecisionTreeClassifier (ccp_alpha=0.0004178989106398837, random_state=42)
scores = cross_val_score(clf_2, X, y, cv=10)

print ("%.2f %.2f" % (scores.mean(), scores.std()))
```

```python
clf = DecisionTreeClassifier (criterion='gini', max_depth=5, random_state=2022)
scores = cross_val_score(clf, X, y, cv=10)

print ("%.2f %.2f" % (scores.mean(), scores.std()))
```
